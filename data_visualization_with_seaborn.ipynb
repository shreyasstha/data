```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

student=pd.read_csv("student.csv")
score=pd.read_csv("score.csv")
classroom=pd.read_csv("classroom.csv")

print(student)
print(score)
print(classroom)
```

```python
sns.countplot(x="age",data=student)

plt.show()
```

```python
sns.scatterplot(x="age", y="student_id",
                data=student,
                hue="gender")
plt.show()
```

```python
hue_colors = {"Female": "pink", "Male": "blue"}
sns.scatterplot(x="age", y="student_id",
                data=student,
                hue="gender", palette = hue_colors)
plt.show()
```

```python
sns.countplot(x="subject" ,data=score)
plt.show()
```
